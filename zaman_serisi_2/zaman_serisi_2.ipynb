```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
plt.style.use("ggplot")
import plotly
plotly.tools.set_credentials_file(username='xx', api_key='xx')
import plotly.plotly as py
import warnings
warnings.filterwarnings("ignore")
import plotly.graph_objs as go 
    
df = pd.read_excel('data/electric_consumption.xlsx',sheet_name='Book1',header=None)
df.columns = ['Date','Consumption']
df.Date = pd.to_datetime(df.Date)
df['Consumption'] = df['Consumption'].str.replace(',', '')
df.tail()
```

```python
trace = go.Scatter(x=list(df.Date),
                   y=list(df.Consumption), line=dict(color='red'))

dat = [trace]
layout = dict(
    title='Energy Consumption of Turkey (MWh)',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

fig = dict(data=dat, layout=layout)
py.iplot(fig)
```

```python
import fbprophet
df = df.rename(columns={'Date': 'ds', 'Consumption': 'y'})
fbp = fbprophet.Prophet()
fbp.fit(df)
```

```python
df_forecast = fbp.make_future_dataframe(periods=24,freq='M')
df_forecast = fbp.predict(df_forecast)
fbp.plot(df_forecast, xlabel = 'Date', ylabel = 'Consumption')
plt.title('Energy Consumption of Turkey (MWh)')
```

```python
fbp.plot(df_forecast, xlabel = 'Date', ylabel = 'Consumption')
plt.vlines([str(date) for date in fbp.changepoints], ymin = 0, ymax= 3e7, color='r',linestyles = 'dashed')
plt.title('Energy Consumption of Turkey (MWh)')
print('Change points:')
print(fbp.changepoints)
```

```python
holiday = pd.DataFrame({
  'ds': pd.to_datetime(['2015-09-01', '2016-07-01']),
  'holiday': 'start',
  'lower_window': 0,
  'upper_window': 3,
})
holiday.head()
```

```python
fbp = fbprophet.Prophet(holidays=holiday)
fbp.fit(df)
df_forecast = fbp.make_future_dataframe(periods=24,freq='M')
df_forecast2 = fbp.predict(df_forecast)
fbp.plot(df_forecast2, xlabel = 'Date', ylabel = 'Consumption')
plt.title('Energy Consumption of Turkey (MWh)')
```
